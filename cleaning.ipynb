```python
import pandas as pd

df = pd.read_csv('Study Case #1 Penjualan Kopi Ratu - Penjualan Kopi 2016-2018.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Mengganti nilai kolom menjadi huruf kapital
df['Nama Depan'] = df['Nama Depan'].str.capitalize()
```

```python
# Menghapus titik pada nilai kolom
df['Nama Depan'] = df['Nama Depan'].str.replace('.', '')
```

```python
# Mengganti nilai kolom menjadi huruf kapital
df['Nama Belakang'] = df['Nama Belakang'].str.capitalize()
```

```python
# Menghapus titik pada nilai kolom
df['Nama Belakang'] = df['Nama Belakang'].str.replace('.', '')
```

```python
# Mengambil hanya nilai numerik
import re
df['Jumlah Pembelian'] = df['Jumlah Pembelian'].apply(lambda x: re.search(r'\d+', x).group())
```

```python
# Mengganti nilai null dengan mean
mean_rating = round(df['Rating'].mean())
df['Rating'].fillna(mean_rating, inplace=True)
```

```python
mode_rating = df['Jasa Pengiriman'].mode().values[0]

# Mengganti nilai null dengan modus
df['Jasa Pengiriman'].fillna(mode_rating, inplace=True)
```

```python
# Menghilangkan koma pada kolom
df['Email'] = df['Email'].str.replace(',', '')
df['Nama Depan'] = df['Nama Depan'].str.replace(',', '')
df['Nama Belakang'] = df['Nama Belakang'].str.replace(',', '')
```

```python
# Mengganti mailx menjadi mail
df['Email'] = df['Email'].str.replace('mailx', 'mail')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Nama Lengkap'] = df['Nama Depan'] + ' ' + df['Nama Belakang']
```

```python
df.drop(['Nama Depan', 'Nama Belakang'], axis=1, inplace=True)
```

```python
data_penjualan = df
```

```python
import pandas as pd

df = pd.read_csv('Study Case #1 Penjualan Kopi Ratu - Produk DB.csv')
```

```python
# DataFrame yang berisi informasi harga
harga_df = df[['SKU', 'Harga']]

# Membuat mapping dari SKU ke Harga
mapping = harga_df.set_index('SKU')['Harga']

# Menggunakan map untuk membuat kolom baru 'Harga Baru'
data_penjualan['Harga'] = data_penjualan['Produk Yang Dibeli'].map(mapping)
```

```python
# DataFrame yang berisi informasi harga
nama_df = df[['SKU', 'Nama']]

# Membuat mapping dari SKU ke Harga
mapping = nama_df.set_index('SKU')['Nama']

# Menggunakan map untuk membuat kolom baru 'Harga Baru'
data_penjualan['Nama Barang'] = data_penjualan['Produk Yang Dibeli'].map(mapping)
```

```python
data_penjualan
```

```python
data_penjualan.info()
```

```python
# Konversi kolom 'Harga' dan 'Jumlah Pembelian' ke tipe data numerik
data_penjualan['Harga'] = data_penjualan['Harga'].str.replace('Rp', '').str.replace(',', '', regex=True).astype(float)
data_penjualan['Jumlah Pembelian'] = data_penjualan['Jumlah Pembelian'].str.replace(',', '', regex=True).astype(int)

# Membuat kolom baru 'Total Harga' dengan hasil perkalian 'Harga' dan 'Jumlah Pembelian'
data_penjualan['Total Harga'] = data_penjualan['Harga'] * data_penjualan['Jumlah Pembelian']
```

```python
data_penjualan.info()
```

```python
# Konversi kolom 'Tanggal Pembelian' ke dalam format datetime jika belum dalam format tersebut
data_penjualan['Tanggal Pembelian'] = pd.to_datetime(data_penjualan['Tanggal Pembelian'])

# Mengurutkan data berdasarkan kolom 'Tanggal Pembelian' secara menaik
data_sorted = data_penjualan.sort_values(by='Tanggal Pembelian')
```
